```python
# 가상의 데이터
import numpy as np
X=2*np.random.rand(100,1) # 균일분포 U(0,1)에서 100*1의 matrix 생성
Y=4+3*X+np.random.randn(100,1) #표준정규분포로부터 100*1의 matrix 생성

# 모든 샘플에 x0=1 추가
Xb=np.c_[np.ones((100,1)),X]

import pandas as pd
pd.DataFrame(Xb)
```

```python
# dot() 메소드는 행렬곱을 위해 사용됨
# np.linalg.inv()는 역행렬을 구함
# Xb.T 는 Xb의 전치행렬(transpose)
theta_best = np.linalg.inv(Xb.T.dot(Xb)).dot(Xb.T).dot(Y)
theta_best
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg

lin_reg.fit(X,Y)
lin_reg.intercept_,lin_reg.coef_
```

```python
# 배치 경사하강법
eta=0.1
n_iteration=1000
N=len(Y)

theta=np.random.randn(2,1) #세타(모수의 추정값,회귀계수의 추정값)의 초기값을 랜덤으로 지정

for _ in range(n_iteration):
    grad=2/N*Xb.T.dot(Xb.dot(theta)-Y)
    theta=theta-eta*grad
    
theta
```

```python
# 확률적 경사 하강법
n_epochs=50
t0,t1=5,50

def learning_schedule(t):
    return t0/(t+t1)

theta=np.random.randn(2,1)

for epoch in range(n_epochs):
    random_idxes=np.random.choice(range(N),size=N,replace=False)
    
    for i,rd_id in enumerate(random_idxes):
        xi,yi=Xb[rd_id:rd_id+1],Y[rd_id]
        grad=2*xi.T.dot(xi.dot(theta)-yi)
        eta=learning_schedule(epoch*N+i)
        theta=theta-eta*grad
theta
```

```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(max_iter = 1000, tol = 1e-3, penalty = None, eta0 = 0.1)
sgd_reg.fit(X, Y.ravel())
sgd_reg.intercept_, sgd_reg.coef_
```

```python
# 미니배치 경사 하강법
n_epochs=1000
t0,t1=5,50

def learning_schedule(t):
    return t0/(t+t1)

theta=np.random.randn(2,1)

batch_size=5
num_batch=N//batch_size

for epocch in range(n_epochs):
    random_idxes=np.random.choice(range(num_batch),size=num_batch,replace=False)
    
    for i,rd_id in enumerate(random_idxes):
        xi, yi = Xb[rd_id*batch_size:(rd_id + 1)*batch_size], Y[rd_id*batch_size:(rd_id + 1)*batch_size]
        grad=2*xi.T.dot(xi.dot(theta)-yi)
        eta=learning_schedule(epoch*N+i)
        theta=theta-eta*grad
        
theta
```
