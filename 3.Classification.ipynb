```python
import pandas as pd
pd.DataFrame(iris["data"])
```

```python
import numpy as np
X=iris["data"][:,3:] #petal widta
X
```

```python
y=(iris["target"]==2).astype(np.int)
y
```

```python
from sklearn.linear_model import LogisticRegression

log_reg=LogisticRegression()
log_reg.fit(X,y)
log_reg.intercept_,log_reg.coef_
```

```python
X_new=np.linspace(0,3,1000).reshape(-1,1) #ì´ˆí•­ì´ 0ì´ê³  1000ë²ˆì§¸ í•­ì´ 3ì¸ ë“±ì°¨ìˆ˜ì—´
y_proba=log_reg.predict_proba(X_new)
X_new,y_proba
```

```python
pd.concat([pd.DataFrame(X_new),pd.DataFrame(y_proba)],axis=1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(X_new,y_proba[:,1],"g-",label="virginica")
plt.plot(X_new,y_proba[:,0],"b--",label="Not virginica")
plt.show()
```
